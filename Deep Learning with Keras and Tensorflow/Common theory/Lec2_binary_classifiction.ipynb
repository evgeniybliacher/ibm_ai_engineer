```python
import numpy as np
```

***In case you would like to plot activation function set the parameter to true***

```python
is_plot_activation_function = True
is_verbose = False
```

```python
def activation_function(x):
    """
    Step activation function.
    """
    return -1 if x < 0 else 1
```

```python
if is_plot_activation_function:
    import matplotlib.pyplot as plt

    x = np.linspace(-10, 10, num=20)
    y = np.array([activation_function(x) for x in x])
    plt.figure(figsize=(8, 4))

    plt.plot(x, y)
    plt.title("Step Activation Function")
    plt.xlabel("Input")
    plt.ylabel("Output")
    plt.grid()
    plt.show()
```

```python
N = 5
upper_bound = 10
delta = 0.1

x1 = np.random.random(N)
x2 = x1 + [np.random.randint(upper_bound)/upper_bound * upper_bound for i in range(N)]
c1 = [x1, x2]
if is_verbose:
    print("x1:", x1, "\nx2:", x2, "\nc1:", c1)
    

x1 = np.random.random(N)
x2 = x1 - [np.random.randint(upper_bound)/upper_bound * upper_bound for i in range(N)] - delta
c2 = [x1, x2]
if is_verbose:
    print("x1:", x1, "\nx2:", x2, "\nc1:", c2)
```

```python
weights = [-0.3, 0.3]
f = [0, 1]
for i in range(N):
  x = np.array([c1[0][i], c1[1][i]])
  y = np.dot(weights, x)
  if is_verbose:
      print("x:", x, "\ny:", y)
  z = activation_function(y)
  if is_verbose:
      print("z:", z)
  if z == 1:
    print("Category 1")
  else:
    print("Category 2")
  
```

```python
if is_plot_activation_function:
    plt.figure(figsize=(8, 4))
    plt.scatter(c1[0][:], c1[1][:], s=N, color='blue', label='Category 1')
    plt.scatter(c2[0][:], c2[1][:], s=N, color='red', label='Category 2')
    plt.title("Data Points with Activation Function")
    plt.xlabel("x1")
    plt.ylabel("x2")
    plt.grid()
    plt.legend()
    plt.plot(f)
```
